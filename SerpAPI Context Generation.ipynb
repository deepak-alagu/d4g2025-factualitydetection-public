```python
df_filtered = df[df["context"].isna() | (df["context"] == "")]
df_filtered.head()
```

```python
from serpapi.google_search import GoogleSearch

ai_answers = []
API_KEY = ""

for index, row in df_filtered.iterrows():
  print(index, " done")
  QUESTION = row['question']
  params = {
      "engine": "google_ai_mode",
      "q": QUESTION,
      "api_key": API_KEY,
      "hl": "en",
      "gl": "us"
  }
  search = GoogleSearch(params)
  try:
    results = search.get_dict()
    text_blocks = results["text_blocks"]
    ai_answer = text_blocks[0]['snippet']
    ai_answers.append(ai_answer)
  except:
    ai_answers.append("")
```

```python
df_filtered["model_context"] = ai_answers
df_filtered.to_csv("serpapi.csv", index=False)
```
